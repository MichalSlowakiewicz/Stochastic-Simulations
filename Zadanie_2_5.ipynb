### Stationary distribution
----
Let us recall a formula to derive stationary distribution:

$$
\pi P = \pi.
$$

Now, we can calculate directly from definiton:

$$
[\pi_1, \pi_2] \begin{pmatrix} 1 - \alpha & \alpha \\ \beta & 1 - \beta \end{pmatrix} = [\pi_1(1 - \alpha) + \pi_2\beta, \quad \pi_1\alpha + \pi_2(1 - \beta)] = [\pi_1, \pi_2]
$$

$$
\alpha \pi_1 = \beta \pi_2, \space\space \pi_1 + \pi_2 = 1
$$

$$
⇒ \pi = \left[ \frac{\beta}{\alpha+\beta}, \frac{\alpha}{\alpha+\beta} \right]
$$


### $P^n$ matrix
----
To calculate this, we will use the hint. Let $p_n = P^n(1, 1) = P(X_n = 1|X_0 = 1)$, then: $p_n = p_{n-1}(1 - α) + (1 - p_{n-1})β$.

Firstly, we will find fixed point of above equation:

$$
p = p(1 - α) + (1 - p)β \space\space \Rightarrow \space\space p = \frac{\beta}{\alpha+\beta}
$$

We will now substract this value from both sides of the equation:

$$
p_n - \frac{\beta}{\alpha + \beta} = p_{n-1}(1 - \alpha - \beta) + \beta - \frac{\beta}{\alpha + \beta} = (1 - \alpha - \beta) \left( p_{n-1} - \frac{\beta}{\alpha + \beta} \right)
$$

Therefore, we obtained a geometric sequence with common ratio $(1 - \alpha - \beta)$, it implies that:
$$p_n - \frac{\beta}{\alpha + \beta} = (1 - \alpha - \beta)^n \left( p_0 - \frac{\beta}{\alpha + \beta} \right)$$

Since $X_0=1$:
$$
p_0 -\frac{\beta}{\alpha + \beta} = 1 - \frac{\beta}{\alpha + \beta} = \frac{\alpha}{\alpha + \beta}
$$
and finally:
$$
P^n(1,1) = p_n = \frac{\beta}{\alpha + \beta} + \frac{\alpha}{\alpha + \beta}(1 - \alpha - \beta)^n
$$

$P^n(1,2)$ can be computed instantly:
$$
P^n(1,2) = 1 - P^n(1,1) =  \frac{\alpha}{\alpha + \beta} - \frac{\alpha}{\alpha + \beta}(1 - \alpha - \beta)^n
$$

To calculate two remaining values, we have to solve the equation as earlier with different initial condition:

Let $p_n = P^n(2, 1) = P(X_n = 1|X_0 = 2)$, then: $p_n = p_{n-1}(1 - α) + (1 - p_{n-1})β$ and $p_0=0$. We obtain:

$$
p_n = \frac{\beta}{\alpha + \beta} + (1 - \alpha - \beta)^n \left( 0 - \frac{\beta}{\alpha + \beta} \right)= \frac{\beta}{\alpha + \beta} - \frac{\beta}{\alpha + \beta}(1 - \alpha - \beta)^n
$$

As a result:

$$
P^n(2,1) = \frac{\beta}{\alpha + \beta} - \frac{\beta}{\alpha + \beta}(1 - \alpha - \beta)^n
$$

$$
P^n(2,2) = \frac{\alpha}{\alpha + \beta} + \frac{\beta}{\alpha + \beta}(1 - \alpha - \beta)^n
$$

Now, we can write:

$$
P^n = \frac{1}{\alpha + \beta} \begin{pmatrix} \beta + \alpha \lambda^n & \alpha - \alpha \lambda^n \\ \beta - \beta \lambda^n & \alpha + \beta \lambda^n \end{pmatrix},
$$
where $\lambda = 1 - \alpha -\beta$.
