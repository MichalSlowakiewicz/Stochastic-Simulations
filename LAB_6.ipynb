```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sampler(a=0.1, b=0.1, n=100, x0=1):
  t = [0]
  traj = [x0]
  for i in range(n):
    if traj[-1] == 1:
      e = np.random.exponential(a)
      t.append(e+t[-1])
      traj.append(2)
    else:
      e = np.random.exponential(b)
      t.append(e+t[-1])
      traj.append(1)
  return t, traj
```

```python
t, traj = sampler()
```

```python
plt.figure(figsize=(18,6))
plt.plot(t, traj, drawstyle='steps-pre')
plt.grid(True, alpha=0.3)
plt.show()
```
