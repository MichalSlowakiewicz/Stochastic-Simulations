## Stationary distribution for a random walk on a graph
---

Let $G = (S, E)$ be a connected undirected graph.
The transition probabilities of a Markov chain are defined as:
$$ P(s, t) = \begin{cases} \frac{1}{\deg(s)} & \text{if } \{s, t\} \in E \\ 0 & \text{if } \{s, t\} \notin E \end{cases} $$

We aim to find the stationary distribution $\pi = (\pi_s)_{s \in S}$, which satisfies the equation $\pi P = \pi$. For any vertex $t \in S$ we may rewrite it as:
$$ \pi_t = \sum_{s \in S} \pi_s P(s, t) $$

Since $P(s, t)$ is non-zero only for neighbours of vertex $t$, the sum reduces to vertices $s$ connected to $t$ by a direct edge:
$$ \pi_t = \sum_{s \sim t} \pi_s \frac{1}{\deg(s)} $$

Now, we see that a reasonable idea is to propose a solution for $\pi_s$, which is proportional to $\deg(s)$. Let $\pi_s = c \cdot \deg(s)$, where $c$ is some constant. Next, we substitute it into our equation:
$$ \sum_{s \sim t} \left( c \cdot \deg(s) \right) \frac{1}{\deg(s)} = \sum_{s \sim t} c $$

Observe that we add the constant $c$ exactly as many times as vertex $t$ has neighbours. The number of neighbours for vertex $t$ is denoted as its degree $\deg(t)$, so:
$$ \sum_{s \sim t} c = c \cdot \deg(t) $$

We obtained our proposed formula for $\pi_t$, which proves that our proposed distribution satisfies the stationary equation.

All that remains is to compute the constant $c$. We know that the sum of probabilities must be equal to one, therefore:
$$ \sum_{s \in S} \pi_s = 1 \implies \sum_{s \in S} c \cdot \deg(s) = 1 \implies c \sum_{s \in S} \deg(s) = 1 \implies c = \frac{1}{\sum_{s \in S} \deg(s)}$$

Now, we utilize a theorem from graph theory which states that the sum of the degrees of all vertices is equal to twice the number of edges in the graph:
$$ \sum_{s \in S} \deg(s) = 2|E| $$

As a result, we obtain $c$:
$$ c = \frac{1}{2|E|} $$

Finally, we acquire the formula for the stationary distribution:
$$ \pi_s = \frac{\deg(s)}{2|E|} $$
