## Limit of the Gibbs distribution
---

We consider the Gibbs distribution on a finite space $X$:
$$ \pi_\beta(x) = \frac{\exp[-\beta H(x)]}{Z_\beta}, \quad \text{where} \quad Z_\beta = \sum_{y \in X} \exp[-\beta H(y)] $$

Our goal is to prove that as $\beta \to \infty$, the distribution converges to a uniform distribution over the set of global minima $X_{min} = \{x \in X : H(x) = \min_{x' \in X} H(x')\}$.

Let $H_{min}$ denote the global minimum value of the function $H(x)$.
We can rewrite the probability $\pi_\beta(x)$ by dividing both the numerator and the denominator by $\exp[-\beta H_{min}]$:

$$ \pi_\beta(x) = \frac{ \frac{\exp[-\beta H(x)]}{\exp[-\beta H_{min}]} }{ \sum_{y \in X} \frac{\exp[-\beta H(y)]}{\exp[-\beta H_{min}]} } = \frac{ \exp[-\beta (H(x) - H_{min})] }{ \sum_{y \in X} \exp[-\beta (H(y) - H_{min})] } $$

Now, let us analyze the behavior of the term $\exp[-\beta (H(x) - H_{min})]$ as $\beta \to \infty$:

1.  **If $x \in X_{min}$:**

    $$H(x) - H_{min} = 0$$
    $$ \lim_{\beta \to \infty} \exp[-\beta \cdot 0] = \exp(0) = 1 $$

2.  **If $x \notin X_{min}$:**


    $$H(x) - H_{min} > 0$$
    $$ \lim_{\beta \to \infty} \exp[-\beta (H(x) - H_{min})] = 0 $$

Now we apply these limits to the entire denominator. The sum splits into two parts: elements that belong to $X_{min}$ and elements that do not.
$$ \lim_{\beta \to \infty} \sum_{y \in X} \exp[-\beta (H(y) - H_{min})] = \sum_{y \in X_{min}} 1 + \sum_{y \notin X_{min}} 0 = |X_{min}| $$

Finally, we substitute these limits back into our modified formula for $\pi_\beta(x)$:

$$ \lim_{\beta \to \infty} \pi_\beta(x) = \begin{cases} \frac{1}{|X_{min}|} & \text{if } x \in X_{min} \\ \frac{0}{|X_{min}|} = 0 & \text{if } x \notin X_{min} \end{cases} $$

This proves that as $\beta \to \infty$, the Gibbs distribution converges precisely to $U(X_{min})$.
